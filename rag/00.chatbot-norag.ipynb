## Building a Chatbot

### Import Libraries

```python
from langchain_openai import ChatOpenAI
from IPython.display import Markdown
```

```python
model = ChatOpenAI(model="qwen3.5:9b", base_url="http://localhost:11434/v1/", api_key="")
```

### Create an LLM Call

This function defines a system prompt to guide the LLM. It takes a user prompt as input, calls the model, and returns the response.

```python
def llm_call(user_prompt: str) -> list[str]:
    prompt = [
      {
        'role': 'system',
        'content': 'You are a Vitis HLS product expert who is very familiar with the user guide and provides the guide to the end user.',
      },
      {
        'role': 'user',
        'content': user_prompt,
      },
    ]
    
    return model.invoke(prompt)
```

Let's start querying the LLM

```python
response = llm_call('How are you?')
print(response.content)
```

Now let's ask questions that may be outside the model's knowledge base. You may notice that the model can produce incorrect or made-up answers (hallucinations).

```python
response1 = llm_call('What Operating Systems are supported by Vitis HLS?')
print(response1.content)
```

```python
response2 = llm_call("Where can I find tutorials and examples for Vitis HLS?")
print(response2.content)
```

```python
response3 = llm_call('Provide the C/C++ code to write a vector add in Vitis HLS?')
Markdown(response3.content)
```

```python
response4 = llm_call("What does the dataflow pragma do?")
print(response4.content)
```

```python
response5 = llm_call("What interfaces does Vitis HLS support?")
print(response5.content)
```

```python
response6 = llm_call("In what chapter can I find documentation about directives?")
print(response6.content)
```
